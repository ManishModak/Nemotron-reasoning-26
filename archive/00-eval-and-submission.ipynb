# 00 Eval And Submission

Default Kaggle notebook for Milestone 1.

This notebook is the prompt-only baseline and debugging path. It:

- runs held-out validation on `train.csv`
- generates `submission.csv` from `test.csv`
- reuses the shared `src/` predictor, parsing, routing, and reporting code

This notebook does **not** produce the final competition artifact.

The final competition path still requires:

- a LoRA adapter
- rank `<= 32`
- `submission.zip`
- evaluation through the official vLLM-based submission pipeline


```python
from pathlib import Path
import json
import sys

ROOT = Path.cwd().resolve()
if not (ROOT / "src").exists():
    for candidate in [ROOT, *ROOT.parents]:
        if (candidate / "src").exists():
            ROOT = candidate
            break

if not (ROOT / "src").exists() and Path("/kaggle/input").exists():
    for nested in Path("/kaggle/input").rglob("src"):
        if nested.is_dir() and (nested.parent / "configs" / "experiments" / "baseline_eval.yaml").exists():
            ROOT = nested.parent
            break

if not (ROOT / "src").exists():
    raise RuntimeError(
        "Could not find repo root with src/ and configs/experiments/baseline_eval.yaml. "
        "Attach the repo dataset to Kaggle and run this notebook from the first cell."
    )

if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

WORKDIR = Path("/kaggle/working") if Path("/kaggle/working").exists() else ROOT
IS_KAGGLE = Path("/kaggle").exists()
ROOT, WORKDIR, IS_KAGGLE
```

```python
import os

from src import load_config
from src.data import load_eval_examples, resolve_dataset_path
from src.eval import build_predictor, evaluate_examples
from src.eval.reporting import export_handoff_bundle
from src.eval.splits import split_examples
from src.prompts import get_prompt_variant, get_prompt_variants
from src.solvers import ConservativeRouter
from src.submission import write_submission_csv

config = load_config(ROOT / "configs/experiments/baseline_eval.yaml")
if IS_KAGGLE and config["predictor"].get("type") == "heuristic":
    config = {
        **config,
        "predictor": {
            **config["predictor"],
            "type": "transformers_kaggle",
        },
    }
    print("Running on Kaggle: overriding predictor.type to transformers_kaggle")

train_path = resolve_dataset_path(
    config["data"].get("train_path"),
    fallback_filename="train.csv",
    auto_discover=config["data"].get("auto_discover", False),
    base_dir=ROOT,
)
test_path = resolve_dataset_path(
    config["data"].get("test_path"),
    fallback_filename="test.csv",
    auto_discover=config["data"].get("auto_discover", False),
    base_dir=ROOT,
)

router = ConservativeRouter(
    confidence_threshold=config["routing"]["confidence_threshold"],
    enabled_families=tuple(config["routing"].get("enabled_families", [])) or None,
)
selected_prompt_names = set(config["prompts"])
prompt_variants = [
    variant
    for variant in get_prompt_variants()
    if variant.name in selected_prompt_names
]
selected_variant = get_prompt_variant(config["selected_prompt_variant"])

candidate_cutlass_paths = [
    Path("/kaggle/usr/lib/notebooks/ryanholbrook/nvidia-utility-script/nvidia_cutlass_dsl/python_packages/"),
    Path("/kaggle/usr/lib/notebooks/ryanholbrook/nvidia_utility_script/nvidia_cutlass_dsl/python_packages/"),
]
cutlass_path_exists = any(path.exists() for path in candidate_cutlass_paths)
train_from_competition = "/kaggle/input/competitions/" in str(train_path).replace('\\', '/')
test_from_competition = "/kaggle/input/competitions/" in str(test_path).replace('\\', '/')
```

```python
print("=== Runtime Diagnostics ===")
print("ROOT:", ROOT)
print("WORKDIR:", WORKDIR)
print("IS_KAGGLE:", IS_KAGGLE)
print("Train file:", train_path)
print("Test file:", test_path)
print("Predictor type:", config["predictor"]["type"])
print("Selected prompt variant:", selected_variant.name)
print("Configured prompt variants:", [variant.name for variant in prompt_variants])
print("Model handle:", config["predictor"].get("model_handle"))
print("Routing enabled:", config["routing"]["enabled"])
print("Routing enabled families:", config["routing"].get("enabled_families", []))
print("TRITON_PTXAS_PATH:", os.environ.get("TRITON_PTXAS_PATH", "<unset before predictor init>"))
print("TRITON_PTXAS_BLACKWELL_PATH:", os.environ.get("TRITON_PTXAS_BLACKWELL_PATH", "<unset before predictor init>"))
print("CUTLASS package path exists:", cutlass_path_exists)
print("Train from competition input:", train_from_competition)
print("Test from competition input:", test_from_competition)
```

## Preflight Warnings

- If `train.csv` or `test.csv` resolve to the repo dataset copy instead of the competition input, stop here and fix the Kaggle inputs before running generation.
- If model initialization or first-token generation later fails because of Triton, PTXAS, or CUTLASS issues, inspect the shared runtime setup in `src/eval/predictors.py` first.
- Local answer extraction and verification have been aligned to the official metric notebook, but Kaggle scoring remains authoritative.


```python
print("Metric alignment: local extraction and answer matching mirror the official metric notebook where possible.")
print("Kaggle scoring remains the final authority.")
predictor = build_predictor(config)
print("Predictor initialized successfully.")
print("TRITON_PTXAS_PATH after init:", os.environ.get("TRITON_PTXAS_PATH", "<unset>"))
print("TRITON_PTXAS_BLACKWELL_PATH after init:", os.environ.get("TRITON_PTXAS_BLACKWELL_PATH", "<unset>"))
```

```python
all_examples = load_eval_examples(train_path)
train_examples, validation_examples = split_examples(
    all_examples,
    validation_ratio=config["data"]["validation_ratio"],
    seed=config["data"]["seed"],
)
selected_examples = validation_examples or all_examples
eval_limit = config["data"].get("eval_limit")
if eval_limit is not None:
    selected_examples = selected_examples[: eval_limit]

eval_result = evaluate_examples(
    selected_examples,
    prompt_variants=prompt_variants,
    predictor=predictor,
    router=router if config["routing"]["enabled"] else None,
    run_name=config["run_name"],
    output_dir=WORKDIR / config["reporting"]["output_dir"],
    failure_sample_size=config["reporting"].get("failure_sample_size", 5),
)
handoff_bundle = export_handoff_bundle(
    eval_result,
    WORKDIR / config["kaggle"]["output_dir"],
    selected_variant=config["selected_prompt_variant"],
)
eval_result.metrics
```

```python
summary_path = eval_result.artifact_paths["summary_md"]
print("=== Validation Summary ===")
print("Validation examples used:", len(selected_examples))
print("Overall accuracy:", eval_result.metrics["overall_accuracy"])
print("Route breakdown:", eval_result.metrics["route_breakdown"])
print("Errors by family:", eval_result.metrics["errors_by_family"])
print("Handoff bundle:", handoff_bundle)
print()
print(summary_path.read_text(encoding="utf-8"))
```

```python
test_examples = load_eval_examples(test_path)

submission_result = evaluate_examples(
    test_examples,
    prompt_variants=[selected_variant],
    predictor=predictor,
    router=router if config["routing"]["enabled"] else None,
    run_name="submission_generation",
    output_dir=WORKDIR / config["kaggle"]["output_dir"],
    failure_sample_size=config["reporting"].get("failure_sample_size", 5),
)
submission_path = write_submission_csv(
    submission_result.predictions,
    WORKDIR / config["submission"]["output_csv"],
)
submission_path
```

```python
note_path = WORKDIR / config["kaggle"]["output_dir"] / "kaggle_run_note.json"
note_path.parent.mkdir(parents=True, exist_ok=True)
note_payload = {
    "status": "eval_and_submission_ready",
    "predictor_type": config["predictor"]["type"],
    "selected_prompt_variant": selected_variant.name,
    "train_path": str(train_path),
    "test_path": str(test_path),
    "submission_csv": str(submission_path),
    "eval_metrics": eval_result.metrics,
    "submission_route_breakdown": submission_result.metrics["route_breakdown"],
    "next_steps": [
        "Validate runtime and memory budget on the full Kaggle hardware path.",
        "Inspect submission formatting before final submit.",
        "Record the Kaggle checkpoint notes in reports/.",
    ],
}
note_path.write_text(json.dumps(note_payload, indent=2), encoding="utf-8")

submission_text = submission_path.read_text(encoding="utf-8")
submission_lines = submission_text.splitlines()
print("=== Submission Summary ===")
print("Submission CSV:", submission_path)
print("Predictions written:", max(len(submission_lines) - 1, 0))
print("First lines:")
for line in submission_lines[:5]:
    print(line)
print()
print("Reminder: this CSV is for baseline/debugging and does not replace the final LoRA submission.zip path.")
print()
print(note_path.read_text(encoding="utf-8"))
```

## What This Notebook Does Not Do

- It does not create `submission.zip`.
- It does not package a LoRA adapter.
- It does not reproduce the final vLLM-based submission pipeline.
- It is the baseline/debugging notebook that should feed into the later final-submission workstream.
